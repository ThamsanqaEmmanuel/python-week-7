```python
import pandas as pd           # For data handling
import matplotlib.pyplot as plt  # For static plots
import seaborn as sns         # For statistical visualizations
import plotly.express as px   # For interactive plots
import geopandas as gpd
```

```python
#Task 1

try:
    df = pd.read_csv("traffic_accidents.csv")
    print("Dataset loaded successfully.")
except Exception as e:
    print("Error loading dataset:", e)
```

```python
#Calling the df to display the table
df.head()
```

```python
#Check for missing values
df.isnull().sum()
```

```python

df['crash_date'] = pd.to_datetime(df['crash_date'])
df['year'] = df['crash_date'].dt.year
```

```python
df.describe()

# Group
accidents_by_year = df.groupby('year').size().reset_index(name='accident_count')
```

```python
#Task 3
# Line chart
plt.figure(figsize=(10, 6))
plt.plot(accidents_by_year['year'], accidents_by_year['accident_count'], marker='o')
plt.title('Traffic Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python

#Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(accidents_by_year['year'], accidents_by_year['accident_count'], color='skyblue')
plt.title('Traffic Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(axis='y')  
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python
#Histogram
plt.figure(figsize=(10, 6))
plt.hist(accidents_by_year['accident_count'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Annual Traffic Accident Counts')
plt.xlabel('Number of Accidents')
plt.ylabel('Frequency (Number of Years)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python

#Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(accidents_by_year['year'], accidents_by_year['accident_count'], color='green')
plt.title('Traffic Accidents by Year (Scatter Plot)')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
